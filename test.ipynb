```python
b = b"hello \xff\xef\xffworld"

# b.decode("utf-8", errors="strict")
# UnicodeDecodeError: 'utf-8' codec can't decode byte 0xff

print(b.decode("utf-8", errors="ignore"))
# 'hello  world'

print(b.decode("utf-8", errors="replace"))
# 'hello ÔøΩ world'

print(b.decode("utf-8", errors="backslashreplace"))
# 'hello \\xff world'
```
